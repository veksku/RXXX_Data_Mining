```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('20newsgroups.csv')
data = data.dropna()
```

Klasifikovati podatke (ciljna kolona je ‘target’) korišćenjem naivnog Bajesovog algoritma i TF matrice. 

Da li je potrebno podeliti podatke na trening i test skup? - Da

```python
X = data['text']
y = data['target']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, stratify=y, random_state=67)

cv = CountVectorizer()
X_train_TF = cv.fit_transform(X_train)
X_test_TF = cv.transform(X_test)

mnb = MultinomialNB()
mnb.fit(X_train_TF, y_train)
```

```python
y_pred = mnb.predict(X_test_TF)
```

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix

print(f'F1 score: {f1_score(y_pred, y_test, average='macro')}')
print(f'Accuracy: {accuracy_score(y_pred, y_test)}')
print(f'Confusion matrix:\n{confusion_matrix(y_pred, y_test)}')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfid = TfidfVectorizer()
X_train_TFID = cv.fit_transform(X_train)
X_test_TFID = cv.transform(X_test)

mnb2 = MultinomialNB()
mnb2.fit(X_train_TFID, y_train)
```

```python
y_pred_2 = mnb2.predict(X_test_TFID)
```

```python
print(f'F1 score: {f1_score(y_pred_2, y_test, average='macro')}')
print(f'Accuracy: {accuracy_score(y_pred_2, y_test)}')
print(f'Confusion matrix:\n{confusion_matrix(y_pred_2, y_test)}')
```

```python
wrong = np.where(y_pred_2 != y_test)[0]
wrong

for i in wrong[:5]:
    print(f'Text:\n "{X_test.iloc[i]}"')
    print(f'True class: {y_test.iloc[i]}')
    print(f'Predicted: {y_pred_2[i]}')
    print("-" * 50)
```

```python
from sklearn.svm import LinearSVC

svm = LinearSVC()

svm.fit(X_train_TFID, y_train)

y_pred_svm = svm.predict(X_test_TFID)
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('moons.csv')
data
```

```python
X = data.values
```

Primenom algoritma K-means pronaći 2 klastera.  

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=2, n_init='auto', random_state=67)
kmeans.fit(X_scaled)

labels = kmeans.predict(X_scaled)

centers = kmeans.cluster_centers_
```

Prikazati grafički rezultate koristeći scatter plot. Obojiti instance na osnovu 
klastera kom pripadaju. Centroide označiti crnom bojom. 

```python
from matplotlib import pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], c="black", marker="X")
```

Primenom algoritma sakupljajućeg hijerarhijskog klasterovanja pronaći 2 
klastera. Koristiti euklidsko rastojanje. 

```python
from sklearn.cluster import AgglomerativeClustering

linkages = ['single', 'complete', 'average']

results = {}

for link in linkages:
    agc = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage=link)

    labels = agc.fit_predict(X)
    results[link] = labels
```

Uporediti rezultate klasterovanja u odnosu na tip veze (min, max, avg), kao i u odnosu na K-means. Da li su rezultati očekivani? 

Prikazati grafički rezultate koristeći scatter plot. Obojiti instance na osnovu klastera kom pripadaju. 

Da li je bila potrebno dodatno pretprocesiranje podataka? Zašto? 

```python
plt.scatter(X[:, 0], X[:, 1], c=results['single'])
plt.scatter(centers[:, 0], centers[:, 1], c="black", marker="X")
plt.title('single')
```

```python
plt.scatter(X[:, 0], X[:, 1], c=results['complete'])
plt.scatter(centers[:, 0], centers[:, 1], c="black", marker="X")
plt.title('complete')
```

```python
plt.scatter(X[:, 0], X[:, 1], c=results['average'])
plt.scatter(centers[:, 0], centers[:, 1], c="black", marker="X")
plt.title('average')
```

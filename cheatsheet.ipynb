```python
import numpy as np
import pandas as pd

data = pd.read_csv('../Datasets/bank.csv')
```

Provera null i izbacivanje null

```python
data.isna().any().any()
data.dropna()
```

Ubacivanje vrednosti umesto null

```python
data['column'].fillna('asd')
data['column'].bbfill()
data['column'].fbfill()
```

Iterative imputation - predvidjanje vrednosti iterativno, sve vrednosti koje imamo koristimo kao model funkcije atributa


```python
data_c = data[['column1', 'column2', 'column3']]

from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

imptr = IterativeImputer(RandomForestRegressor(), max_iter=100, tol=0.01, random_state=67)
data2 = pd.DataFrame(imptr.transform(data_c), columns = data_c.columns)
```

Normalizacija

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
data = scaler.fit_transform(data)
```

Stablo odlucivanja

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X = data.drop['target']
y = data['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, stratify=y, random_state=67)
dtc = DecisionTreeClassifier()

dtc.fit(X_train, y_train)
```

Slucajne sume

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=2)
random_forest.fit(X_train, y_train)
```

K najblizih suseda (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

Ansambl

```python
from sklearn.ensemble import BaggingClassifier

baggingKnn = BaggingClassifier(estimator=KNeighborsClassifier(), n_estimators=20)
baggingKnn.fit(X_train, y_train)
```

Kategoricki naivni bajes i njegov enkoder

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
oe.fit(X_train)

X_train_ordinal = oe.transform(X_train)
X_test_ordinal = oe.transform(X_test)

from sklearn.naive_bayes import CategoricalNB

cnb = CategoricalNB()
cnb.fit(X_train_ordinal, y_train)

y_train_pred = cnb.predict(X_train_ordinal)
y_test_pred = cnb.predict(X_test_ordinal)
```

Multinomijalni bajes i TF matrica

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer()
X_train_TF = dv.fit_transform(X_train)
X_test_TF = dv.transform(X_test)

from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()
mnb.fit(X_train_TF, y_train)

y_test_pred = mnb.predict(X_test_TF)
```

Za tekst countvectorizer i tfid

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

cv = CountVectorizer() # slicno X_train_cv i X_test_cv

tfid = TfidfVectorizer()
X_train_tfid = tfid.fit_transform(X_train)
X_test_tfid = tfid.fit_transform(X_test)

mnbtfid = MultinomialNB()
mnbtfid.fit(X_train, y_train)

y_test_pred_tfid = mnbtfid.predict(X_test_tfid)
```

Matrica konfuzije i accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score

confusion_matrix(y_train, y_train_pred)
confusion_matrix(y_test, y_test_pred)
accuracy_score(y_train, y_train_pred)
accuracy_score(y_test, y_test_pred)
```

PCA


```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_train)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

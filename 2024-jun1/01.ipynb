```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

Učitati skup podataka 20newsgroups.

```python
data = pd.read_csv('20newsgroups.csv')
```

```python
data.head()
```

Koliko ima različitih klasa?

```python
len(np.unique(data['target']))
```

Ako je potrebno, podeliti skup podataka na deo za obučavanje i deo za testiranje modela.

```python
X = data['text']
y = data['target']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, stratify=y, random_state = 67)
```

Da li je skup balansiran? Ako je potrebno, primeniti slučajno uzorkovanje skupa.

```python
np.bincount(y)


#slucajno uzorkovanje

# from sklearn.utils import resample

# ili

# X_train_frac = X_train.sample(frac=0.2, random_state=67)
# X_test_frac = X_test.sample(frac=0.2, random_state=67)
# y_train_frac = y_train.sample(frac=0.2, random_state=67)
# y_test_frac = y_test.sample(frac=0.2, random_state=67)
```

Pretvoriti tekstualne podatke u matricu terma u dokumentima (term frequency, TF matrica).

```python
cv = CountVectorizer()
X_train_tf = cv.fit_transform(X_train)
X_test_tf = cv.transform(X_test)
```

Primeniti Naivni Bajesov algoritam za klasifikaciju.

```python
model = MultinomialNB()
model.fit(X_train_tf, y_train)
```

```python
y_pred = model.predict(X_test_tf)
```

Proceniti kvalitet dobijenog modela koristeći matricu konfuzije i tačnost (accuracy) modela. 

Da li je tačnost adekvatna metrika za ovaj skup podataka?

```python
print(f'Matrica konfuzije: \n{confusion_matrix(y_pred, y_test)}')
print(f'Accuracy score: {accuracy_score(y_pred, y_test)}')
```

Pretvoriti TF matricu u TF-IDF i ponoviti ceo postupak. Uporediti novi model sa starim.

```python
tfid = TfidfVectorizer()

X_train_tfid = tfid.fit_transform(X_train)
X_test_tfid = tfid.transform(X_test)
```

```python
model.fit(X_train_tfid, y_train)
```

```python
y_pred = model.predict(X_test_tfid)
```

```python
print(f'Matrica konfuzije: \n{confusion_matrix(y_pred, y_test)}')
print(f'Accuracy score: {accuracy_score(y_pred, y_test)}')
```
